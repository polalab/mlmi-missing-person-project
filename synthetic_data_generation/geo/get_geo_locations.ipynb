```python
import pandas as pd
import matplotlib.pyplot as plt
import osmnx as ox
```

```python
place = "Scotland, UK", 
sc_df = ox.features.features_from_address(
    place, {'building': True}, dist=100000)

valid_buildings = [
    'residential', 'train_station', 'commercial', 'hotel',
    'college', 'supermarket', 'public', 'religious', 'hospital'
]
sc_df = sc_df[sc_df['building'].isin(valid_buildings)]
sc_df = sc_df[sc_df['addr:street'].isna()==False]
sc_df[['building','addr:street', 'addr:postcode', 'addr:housenumber', 'addr:city']]
```
