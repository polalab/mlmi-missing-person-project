```python
def parse_questions(text):
    """
    Extract explanations that come after "Yes -" or "No -" in numbered questions.
    Returns a dictionary with question numbers as keys and explanations as values.
    """
    # Pattern to match: number. question - Yes/No - explanation
    pattern = r'(\d+)\.\s+[^-]+-\s+(?:Yes|No)\s+-\s+([^0-9]+?)(?=\s*\d+\.|$)'
    
    matches = re.findall(pattern, text, re.DOTALL)
    
    result = {}
    for match in matches:
        number, explanation = match
        # Clean up the explanation
        explanation = explanation.strip()
        result[f"q_{number}"] = explanation
    
    return result
parse_questions("Vulnerability 1. Is there any identified risk of suicide? - Yes - MP is known to frequent train stations in an attempt to commit suicide and also to drink antifreeze. 2. Is criminality suspected to be a factor in the disappearance? - No 3. Is the Missing Person vulnerable due to age, infirmity or other similar factor? - Yes - MP is subject to a Compulsory Treatment Order and has poor mental health. 4. Are there any known adverse effects if prescribed medication is not available to the Missing Person? - No 5. Does the Missing Person have dementia, medical or mental health conditions, physical illnesses or disabilities? - Yes - MP has poor mental health and has thoughts of suicide. 6. Does the Missing Person have problems interacting safely with others when finding themselves in unfamiliar circumstances? - No 7. Is there a dependency on drugs, alcohol, medication or other substances? - No 8. Is the Missing Person on the Child Protection Register? - No 9. Do the current/previous weather conditions present additional risk? Consider all circumstances including age and clothing. - No Influences 10. Are there family/relationship problems or recent history of family conflict and/or abuse? - No 11. Is the Missing Person a victim or perpetrator of domestic violence? - No 12. Is there an ongoing personal issue linked to racial, sexual, homophobic, the local community or any cultural issues? - No 13. Was the Missing Person involved in a violent and/or hate crime incident prior to disappearance - No 14. Are there any school, college, university, employment or financial problems? - No 15. Is forced marriage or honour based abuse an issue? - No 16. Is the Missing Person the victim of sexual exploitation, human trafficking or prostitution? If so, is going missing likely to place the Missing Person at risk of considerable harm. - No Behaviour 17. Are the circumstances of going missing different from normal behaviour patterns? - No 18. Is there any known reason for the Missing Person to go missing? - Yes - MP frequnetly absconds from Ward X. 19. Are there any indications that preparations have been made for absence? - No 20. What was the Missing Person intending to do when last seen? Did they fail to complete their intentions? -  - Mps intentions are unknown however she failed to return at her allocated time. 21. Has the Missing Person disappeared previously and were they exposed to harm on such occassions? - Yes - MP last absocnded and was declared missing on 08/04/2022. 22. Is the missing person a risk to others? And if so in what way? - No 23. Are their other unlisted factors which the officer or supervisor considers relevant in the assessment of risk? - No Mental 24. Does the Missing Person have a mental health condition (excluding dementia)?   - No 25. Does this Missing Person have dementia or a specific progressive neurological disorder?  - No In consideration of the above factors, their likelihood and their seriousness, what level of risk do you consider to be appropriate? Any lower risk grading would be adequate but may not be appropriate. Medium Any other comments you may have: MP frequently absconds from hospital and this is considered normal behaviour for her. Risk Assessment Log 06/05/2022 1 - Expresssed sucidial ideation and consumed anti-freeze whilst missing. Enquiries and investigation led as HR MP  - Brian Brooks (1456789) - 07:18:18 05/05/2022 2 - MP frequently absconds from hospital and this is considered normal behaviour for her.  - Noah Wright (1987654) - 13:45:50	")
    
```
