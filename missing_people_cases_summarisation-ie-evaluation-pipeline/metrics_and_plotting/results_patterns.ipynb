```python
denom = pattern_results['exact_matches'] + pattern_results['total_inserted'] + pattern_results['partial_matches']+ pattern_results['cosine_sim_truth']
pattern_results['exact_precision'] = pattern_results['exact_matches'] / denom
pattern_results['part_precision'] =  (pattern_results['exact_matches'] + pattern_results['partial_matches']) / denom
pattern_results['part_cos_precision'] = (pattern_results['exact_matches'] +  pattern_results['partial_matches'] + pattern_results['cosine_sim_truth']  )/ denom

recall_denom =  (pattern_results['exact_matches']  + pattern_results['missing_from_truth'] + pattern_results['partial_matches'] +pattern_results['cosine_sim_truth'] )
pattern_results['recall'] = pattern_results['exact_matches'] / recall_denom
pattern_results['part_recall'] = (pattern_results['partial_matches']  + pattern_results['exact_matches'])/ recall_denom
pattern_results['part_cos_recall'] =  (pattern_results['partial_matches']  + pattern_results['exact_matches'] +pattern_results['cosine_sim_truth'] )/ recall_denom
```

```python
import matplotlib.lines as mlines

def create_professional_boxplot(df, columns, title, subtitle):

    # Reshape the data for Seaborn
    df_melted = df[columns].melt(var_name='Metric', value_name='Value')

    # Create a new column to group metrics for coloring
    df_melted['Metric Type'] = df_melted['Metric'].apply(
        lambda x: 'Precision' if 'precision' in x.lower() else 'Recall'
    )

    df_melted['Metric'] = df_melted['Metric'].str.replace('_', ' ').str.title()

    sns.set_theme(style="ticks")
    fig, ax = plt.subplots(figsize=(18, 10))

    palette = {"Precision": "#2c7bb6", "Recall": "#fdae61"}

    # Boxplot as the main layer
    sns.boxplot(
        x='Metric',
        y='Value',
        hue='Metric Type',
        data=df_melted,
        palette=palette,
        showmeans=True,
        meanprops={"marker": "D", "markerfacecolor": "white", "markeredgecolor": "black", "markersize": "8"},
        dodge=False,
        ax=ax
    )

    # --- Annotation Loop ---
    # Calculate offset for placing text
    ax.set_ylim(top=ax.get_ylim()[1] * 1.1)  # Increase y-axis limit to make space for text

    # Get the unique metrics in the order they appear on the plot
    ordered_metrics = [tick.get_text() for tick in ax.get_xticklabels()]

    for i, metric in enumerate(ordered_metrics):
        # Filter data for the current metric
        metric_data = df_melted[df_melted['Metric'] == metric]['Value']
        
        # Calculate mean and standard deviation
        mean_val = metric_data.mean()
        std_val = metric_data.std()
        
        # Define the annotation text
        annotation_text = f"Mean: {mean_val:.2f}\nStd: {std_val:.2f}"
        
        # Position the text above the box plot whiskers
        y_pos = metric_data.max() * 1.02 # Place it slightly above the max value
        
        ax.text(i, y_pos, annotation_text,
                horizontalalignment='center',
                size=16,
                color='black',
                )


    # --- Final Touches ---
    # fig.suptitle(title, fontsize=22, weight='bold', ha='center', y=0.98)
    # ax.set_title(subtitle, fontsize=16,  color='dimgray',   ha='center', y=0.97)

    # Improve axis labels and ticks
    ax.set_xlabel("Metric", fontsize=18, weight='bold', labelpad=15)
    ax.set_ylabel("Score", fontsize=18, weight='bold', labelpad=15)
    plt.xticks(fontsize=18, rotation=15, ha='right') # Rotate for better readability
    plt.yticks(fontsize=18)

    # Add a light grid for better readability
    ax.yaxis.grid(True, linestyle='--', which='major', color='lightgrey', alpha=0.7)

    
    handles, _ = ax.get_legend_handles_labels()
    
    # Create a custom handle for the mean marker
    mean_handle = mlines.Line2D([], [], color='black', marker='D', linestyle='None',
                              markersize=8, markerfacecolor='white', markeredgecolor='black', label='Mean')
    median_handle = mlines.Line2D([], [], color='black', linestyle='-', linewidth=1)

    # Combine handles and create new labels
    combined_handles = handles + [mean_handle, median_handle]
    labels = ['Precision', 'Recall', 'Mean', 'Median']

    # Add and style the legend, moving it slightly up and out
    ax.legend(handles=combined_handles, labels=labels,
              title='Legend', fontsize=16, title_fontsize=14,
              loc='upper left', bbox_to_anchor=(1.01, 1.0)) # Move legend outside plot

    # Clean up the visual clutter
    sns.despine(trim=True, left=True)
    plt.tight_layout(rect=[0, 0, 1, 0.95])

    plt.savefig("plots/patterns_boxplot.svg", bbox_inches='tight')
    plt.show()


# --- 3. Generate the Plot ---
metrics_cols = ['exact_precision', 'part_precision', 'part_cos_precision', 'recall', 'part_recall', 'part_cos_recall']
plot_title = "Distribution of Precision and Recall Metrics \n Assosiation Network"
plot_subtitle = "Association Network Entities"

# Call the updated function
create_professional_boxplot(pattern_results, metrics_cols, plot_title, plot_subtitle)
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import linregress

df = pattern_results.copy()
y_axis_col = 'len_summ'
x_axis_col = 'len_truth'

# --- Calculate Regression Statistics ---
# Use scipy.stats.linregress to get detailed regression results
# It's important to drop any NaN values for the calculation to work
clean_df = df[[x_axis_col, y_axis_col]].dropna()
slope, intercept, r_value, p_value, std_err = linregress(
    clean_df[x_axis_col],
    clean_df[y_axis_col]
)
r_squared = r_value**2

# --- Calculate Descriptive Statistics ---
x_stats = df[x_axis_col].describe()
y_stats = df[y_axis_col].describe()

# --- Plotting ---
sns.set_theme(style="whitegrid")
plt.figure(figsize=(12, 7))

# Create the scatter plot
scatter_plot = sns.regplot(
    x=x_axis_col,
    y=y_axis_col,
    data=df,
    scatter_kws={'alpha': 0.6, 's': 80, 'color': '#4A90E2'}, # Style the points
    line_kws={'color': 'red', 'linestyle': '--'} # Style the regression line
)

# --- Labels & Styling ---
plt.xlabel('Number of injected patterns in a case', fontsize=18)
plt.ylabel('Number inferred patterns in a case', fontsize=18)

# --- Add Descriptive Statistics to the plot ---
# Create a formatted multiline string with descriptive statistics for both axes
stats_text = f"""No. injected patterns:
Mean: {x_stats['mean']:.2f}
Median: {x_stats['50%']:.2f}
Std: {x_stats['std']:.2f}
Min: {x_stats['min']:.0f}
Max: {x_stats['max']:.0f}

No. inferred patterns:
Mean: {y_stats['mean']:.2f}
Median: {y_stats['50%']:.2f}
Std: {y_stats['std']:.2f}
Min: {y_stats['min']:.0f}
Max: {y_stats['max']:.0f}

Slope = {slope:.4f}"""

# Place the text box on the plot for easy reference
plt.text(0.75, 0.75, stats_text, transform=plt.gca().transAxes, fontsize=12,
         verticalalignment='top', bbox=dict(boxstyle='round,pad=0.5', fc='wheat', alpha=0.8))

# Save the figure with a tight bounding box
plt.tight_layout()
plt.savefig("plots/patterns_length_rel.svg", bbox_inches='tight')

plt.show()

# --- Print Descriptive and Regression Statistics to Console ---
print("="*40)
print("      Regression Statistics")
print("="*40)
print(f"Slope: {slope:.4f}")
print(f"Intercept: {intercept:.4f}")
print(f"R-value (Correlation Coefficient): {r_value:.4f}")
print(f"R-squared: {r_squared:.4f}")
print(f"P-value: {p_value:.4f}")
print(f"Standard Error of the Estimate: {std_err:.4f}")
print("\n" + "="*40 + "\n")

print("Descriptive Statistics for X-Axis:\n")
print(df[x_axis_col].describe().round(3))
print("\n" + "="*40 + "\n")
print("Descriptive Statistics for Y-Axis:\n")
print(df[y_axis_col].describe().round(3))
```
